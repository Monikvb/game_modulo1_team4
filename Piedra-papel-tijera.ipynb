##Juego de Piedra, Papel o Tijera ✏️ ##

Este es un juego diseñado para dos jugadores, uno de los cuales es el jugador humano y el otro es la máquina. El objetivo del juego es acumular tres victorias antes que el oponente.

**Cómo Jugar:**
1. Al inicio del juego, se le pedirá al jugador que ingrese su nombre para identificarlo durante el juego.
2. En cada ronda, el jugador humano debe elegir entre las opciones de `Piedra, Papel o Tijera`.
3. La máquina elegirá aleatoriamente una de las opciones.
4. Las reglas del juego son simples:
   - La piedra vence a las tijeras.
   - Las tijeras vencen al papel.
   - El papel vence a la piedra.
   - Si ambos jugadores eligen la misma opción, la ronda termina en empate.
5. El juego continuará hasta que uno de los jugadores alcance tres victorias. Además, hay un límite de intentos inválidos para el jugador humano. Si este límite se alcanza, el juego se terminará.

**Estructura del Código:**
- El juego utiliza la biblioteca `random` para que la máquina haga su elección aleatoria.
- El juego se implementa utilizando un bucle `while` y unas condiciones de `if/else`.
- Se utilizan contadores para registrar las victorias de cada jugador y los intentos inválidos.
- Se verifica la validez de la entrada del jugador humano para evitar errores durante el juego.

¡Diviértete jugando al Piedra, Papel o Tijera!

```python
# importo el modulo random y este genera aleatoriedad
import random 

print('Bienvenido al juego!')
print('Este juego consiste en escoger una de las 3 opciones y al azar la maquina escogerá su mejor opción para competir y poder ganar. Podrás ganar con 3 puntos. ¡Suerte!')

# pedir los nombre de cada jugador y asignarle una variable para cuando gane
name_player1 = input('Hola jugador/a #1, nombra tu avatar:')
name_player2 = 'ROBOT'

# Inicializo las partidas, y va sumando las ganadas por jugador - debo sumar cada round del jugador 
winings_player1 = 0
winings_player2 = 0

# Opciones del juego
options_game = ['piedra', 'papel', 'tijera']

# Contador de intentos inválidos
intentos_invalidos = 0
 
# Sigue el bucle hasta que gane uno con 3 puntos de 5
while winings_player1 < 3 and winings_player2 < 3:
    
    game_player1 = input('Hola ' + name_player1 + ' Escoge piedra, papel o tijera').lower()
    
    if game_player1 not in options_game:
        intentos_invalidos += 1
        # Si se alcanzan tres intentos inválidos, detener el juego
        if intentos_invalidos == 3:
            print('Has superado el límite de intentos inválidos. Fin del juego.')
            break
        
        print('Opción no válida!')
        continue  # Continuar con la próxima iteración del bucle sin evaluar el resultado de esta ronda
    
    intentos_invalidos = 0  # Reiniciar el contador de intentos inválidos
    
    game_player2 = random.choice(options_game) # Aqui la función ejecuta la aleatoriedad
    print(f'PLAYER_1 {game_player1} y PLAYER_2 {game_player2}') # un print para mostrar que escogío cada uno
    
    if game_player1 == game_player2 : # cada uno de los intentos, irá hasta la condición que sea correcta
        print("empate")

    if game_player1 == "piedra" and game_player2 == "papel" : 
        print("Gana", name_player2)
        winings_player2 += 1 # sumo lo intentos del jugador en cada ciclo

    if game_player1 == "piedra" and game_player2 == "tijera" : 
        print("Gana", name_player1)
        winings_player1 += 1

    if game_player1 == "papel" and game_player2 == "tijera" : 
        print("Gana", name_player2)
        winings_player2 += 1

    if game_player1 == "papel" and game_player2 == "piedra" : 
        print("Gana", name_player1)
        winings_player1 += 1    

    if game_player1 == "tijera" and game_player2 == "piedra" : 
        print("Gana", name_player2)
        winings_player2 += 1

    if game_player1 == "tijera" and game_player2 == "papel" : 
        print("Gana", name_player1)
        winings_player1 += 1
        
# Se acaba el juego si se cumple la condición    
if winings_player1 > winings_player2:    
    print("Haz ganado ", name_player1)

else:
    print("Ha ganado el", name_player2)
```
