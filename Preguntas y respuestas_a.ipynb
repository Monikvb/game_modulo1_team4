## Juego de Preguntas y Respuestas sobre Geografía ✏️ ##

    Este juego está diseñado para dos usuarios, en este caso uno de ellos es la máquina y el otro es la persona. La máquina le formulará preguntas sobre geografía al usuario, el cual debe contestar escribiendo la respuesta. Cada pregunta correcta suma 1 punto y las incorrectas 0 puntos. Para ganar, la persona deberá contestar correctamente 3 de las 5 preguntas que le formulará la máquina.


Cómo jugar:

1. La máquina formulará una pregunta:
2. El jugador deberá contestar introduciendo la respuesta.
3. Si la respuesta es correcta, se muestra un mensaje de acierto y se suma un punto.
4. Si es incorrecto, se mostrará un mensaje de incorrecto y se descuenta un intento.
5. El juego finalizará cuando el jugador obtenga 3 puntos y gane, o agote los 5 intentos y pierda.


Estructura del código:

1. Se importa la librería random para generar aletoriedad al formular las preguntas.
2. Se establece un diccionario donde las claves serán las preguntas, y las respuestas serán los valores.
3. Creamos una lista con las preguntas del diccionario.
4. Creamos los contadores para llevar el recuento de los intentos y de los puntos, y establecemos el número de oportunidades y los puntos para ganar.
5. Utilizamos un bucle for para que iterase sobre las preguntas mientras se cumpliesen las condiciones establecidas.
6. Dentro del while utilizamos un if - else para comprobar si la respuesta del jugador es correcta o no.
7. Por último, utilizamos un if - else para comprobar la puntuación obtenida y determinar si el jugador gana o pierde.

```python

import random

print("¡Bienvenid@ al juego de preguntas y respuestas sobre geografía!")
print("------------------------------------------------")
print("Para ganar tendrás que contestar correctamente, 3 de las 5 preguntas que se te formularán")
print("------------------------------------------------")


preguntas_respuestas = {
    "¿Cuál es la capital de China?": "pekin",
    "¿Cuál es el río más grande de África?": "nilo",
    "¿En qué comunidad autónoma se habla el Catalán?": "cataluña",
    "¿Quien descubrió America?": "cristobal colon",
    "¿Cuál es el río más grande de España?": "tajo",
    "¿Cuál es la comunidad autónoma más pequeña de España?": "ceuta",
    "¿Cual es la capital de España?" : "madrid",
    "¿La Sagrada Familia en qué ciudad de España se encuentra?": "barcelona",
    "¿Dónde nació Pablo Picasso?": "malaga",
    "¿Dónde desemboca el río Amazonas?": "oceano atlantico",
    "¿Qué ciudad se conoce como la ciudad de la luz?": "parís",
    "¿Cuál es el estado más pequeño del mundo?": "vaticano",
    "¿Cuál es el pico más grande del mundo?": "Everest",
    "¿Cuántas comunidades autónomas hay en España?": "diecisiete",
    "¿Cuál es el pico más alto de España?": "teide"
}


# Convertimos las claves del diccionario en una lista para poder usar random.choice()
preguntas = list(preguntas_respuestas.keys())

# Lista para llevar un registro de las preguntas que ya se han hecho
preguntas_hechas = []

# Variable para contar los puntos del jugador
intentos = 0
puntos = 0
oportunidades = 5
numero_puntos_para_ganar = 3


# Bucle para continuar preguntando mientras ambas sean verdad o True
while puntos < numero_puntos_para_ganar and intentos < oportunidades:
    
    # Hacemos la selección aleatoria de la pregunta
    pregunta = random.choice(preguntas)

    # Verificamos si la pregunta ya se ha hecho
    while pregunta not in preguntas_hechas:
        pregunta = random.choice(preguntas)
        preguntas_hechas.append(pregunta)  # Agregamos la pregunta actual a la lista de preguntas hechas
    
    
    # Mostramos la pregunta al jugador y obtenemos su respuesta
    respuesta_jugador = input(pregunta + "\n: ").lower()

    # Obtenemos la respuesta correcta
    respuesta_correcta = preguntas_respuestas[pregunta]

    # Comparamos la respuesta del jugador con la respuesta correcta
    if respuesta_jugador == respuesta_correcta:
        intentos += 1
        puntos += 1 
        print(pregunta)
        print(f"¡Correctol! es: {respuesta_correcta}")
        print("------------------------------------------------")  
    else:
        intentos += 1
        print(pregunta)
        print(f"¡Oh, has fallado! La respuesta es: {respuesta_correcta}")
        print("------------------------------------------------")

if puntos >= numero_puntos_para_ganar:    
    print(f"Has ganado con: {puntos} puntos")

else:
    print(f"Has obtenido {puntos} puntos y tenias {intentos} intentos. Has perdido, suerte la próxima vez.")
        
```
